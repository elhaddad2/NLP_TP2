```python
texte = '''ربما كانت أحد أھم التطورات التي قامت بھا الریاضیات العربیة التي بدأت في ھذا الوقت بعمل الخوارزمي وھي بدایات الجبر، ومن المھم فھم كیف كانت ھذه الفكرة الجدیدة مھمة، فقد كانت خطوة ثوریة بعیدا عن المفھوم الیوناني للریاضیات التي ھي في جوھرھا ھندسة، الجبر كان نظریة موحدة تتیح الأعداد الكسریة والأعداد اللا كسریة، والمقادیر الھندسیة وغیرھا، أن تتعامل على أنھا أجسام جبریة، وأعطت الریاضیات ككل مسارا جدیدا للتطور بمفھوم أوسع بكثیر من الذي كان موجودا من قبل، وقدم وسیلة للتنمیة في ھذا الموضوع مستقبلا. وجانب آخر مھم لإدخال أفكار الجبر وھو أنھ سمح بتطبیق الریاضیات على نفسھا بطریقة لم تحدث من قبل.'''
print(texte)
```

# **Work Token**

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
from nltk import pos_tag
from nltk import RegexpParser
from nltk.tokenize import sent_tokenize

tokens = nltk.word_tokenize(texte)
print("After Split:",tokens)
```

```python
sent=sent_tokenize(texte)
print("After sentence token:",sent)
```

# **Bag-of-Words**

```python
from keras.preprocessing.text import Tokenizer
 
# using tokenizer 
model = Tokenizer()
model.fit_on_texts(sent)
 
#print keys 

print(f'Key : {list(model.word_index.keys())}')
 
#create bag of words representation 
rep = model.texts_to_matrix(sent, mode='count')
print()
print(rep)
```

# **Term Frequency**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf  = TfidfVectorizer()
corpus = tfidf.fit_transform(sent)
print(tfidf.get_feature_names())
print(corpus)
```
